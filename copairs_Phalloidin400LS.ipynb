Copairs analysis which includes the comparison of standard CP data with the Phalloidin 400LS data.

* **Normalization:** Negcon normalization 
* **mAP calculation:** The following notebook includes the data from mAP calculated as difference to controls as well as the mAP calculated as difference to other treatments. 

```python
import pandas as pd
```

#### Reading the dataframes

#### mAP calculated as difference to controls 

##### Standard CP data

```python
#Reading the results dataframe of standard plate
result_std = pd.read_csv('copairs_csv\\Result_Negcon_wrt_Controls_StandardCP.csv')
```

```python
# A new dataframe is created to have a mean number of cell count 
mean_cellcount_std = result_std.groupby('Metadata_BRD ID')['Metadata_Count_Cells'].mean().to_frame() # Mean of replicates of 'Metadata_Count_Cells' for every compound is taken 
mean_cellcount_std = mean_cellcount_std.reset_index()
mean_cellcount_std = mean_cellcount_std.rename(columns={'Metadata_BRD ID':'BRD ID', 'Metadata_Count_Cells':'Metadata_Count_Cells_Std'})
```

##### Phalloidin 400 LS - long stoke shifted actin 

```python
#Reading the results dataframe of Phalliodin 400 LS plate
result_act = pd.read_csv('copairs_csv\\Result_Negcon_wrt_Controls_Phalloidin400LS.csv')
```

```python
# A new dataframe is created to have a mean number of cell count 
mean_cellcount_act = result_act.groupby('Metadata_BRD ID')['Metadata_Count_Cells'].mean().to_frame() # Mean of replicates of 'Metadata_Count_Cells' for every compound is taken 
mean_cellcount_act = mean_cellcount_act.reset_index()
mean_cellcount_act = mean_cellcount_act.rename(columns={'Metadata_BRD ID':'BRD ID', 'Metadata_Count_Cells':'Metadata_Count_Cells_act'})
```

```python
#Reading the combined dataframe which has the precision values of all the plates (standard, Mito, Phalloidin); mAP - difference to controls 
combined_moa_df = pd.read_csv('copairs_csv\\PrecisionValues_with_MoA_allplates_Negcon_wrt_Controls.csv') 
```

##### Merging the cell count details with the combined dataframe

```python

#Adding the cell count details of standard plate to the final combined df 
combined_moa_cellcount_df = pd.merge(combined_moa_df,mean_cellcount_std, on ='BRD ID')

#Adding the cell count details of actin plate to the final combined df 
combined_moa_cellcount_df = pd.merge(combined_moa_cellcount_df,mean_cellcount_act, on='BRD ID')

#Saving the combined dataframe 
combined_moa_cellcount_df.to_csv('copairs_csv\\PrecisionValues_with_MoA_allplates_cellcount_Negcon_wrt_Controls.csv')
```

```python
combined_moa_cellcount_df
```

#### Normalizing the cell count 

```python
 #Normalizing the cell count for easier plotting 
 combined_moa_cellcount_df['Metadata_Count_Cells_act_norm'] = combined_moa_cellcount_df['Metadata_Count_Cells_act']/100
 combined_moa_cellcount_df['Metadata_Count_Cells_Std_norm'] = combined_moa_cellcount_df['Metadata_Count_Cells_Std']/100
 combined_moa_cellcount_df.to_csv('copairs_csv\\PrecisionValues_with_MoA_allplates_cellcount_Negcon_wrt_Controls.csv')
```
